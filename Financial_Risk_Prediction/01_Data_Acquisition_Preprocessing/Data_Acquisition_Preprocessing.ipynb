````python
import yfinance as yf
import pandas as pd
````

#### **2Îã®Í≥Ñ: Îç∞Ïù¥ÌÑ∞ ÏàòÏßë ÎåÄÏÉÅ Í∏∞ÏóÖ ÏÑ†Ï†ï**
ÌäπÏ†ï ÏÇ∞ÏóÖÏù¥ÎÇò Íµ≠Í∞ÄÏùò Í∏∞ÏóÖÎì§ÏùÑ ÏÑ†Ï†ïÌïòÏó¨ Î∂ÑÏÑù Î≤îÏúÑÎ•º Ï¢ÅÌûàÎäî Í≤ÉÏù¥ Ï¢ãÏäµÎãàÎã§. ÏòàÎ•º Îì§Ïñ¥, ÎØ∏Íµ≠Ïùò Í∏∞Ïà†Ï£ºÎ•º Î∂ÑÏÑùÌïòÍ≥† Ïã∂Îã§Î©¥ Îã§ÏùåÍ≥º Í∞ôÏù¥ Ìã∞Ïª§(Ticker)Î•º Î¶¨Ïä§Ìä∏Î°ú ÎßåÎì≠ÎãàÎã§.

````python
# Î∂ÑÏÑù ÎåÄÏÉÅ Í∏∞ÏóÖ Ìã∞Ïª§ Î¶¨Ïä§Ìä∏ (Ïòà: ÎØ∏Íµ≠ Í∏∞Ïà†Ï£º)
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']
````

> ####  üîç **Ï∂îÍ∞Ä ÌïôÏäµ: Ìã∞Ïª§ Ïã¨Î≥ºÏùò ÏùòÎØ∏**
> AAPL: Apple Inc. (Ïï†Ìîå) - iPhone, iPad, Mac Îì±ÏúºÎ°ú Ïú†Î™ÖÌïú Í∏∞Ïà† Í∏∞ÏóÖ
>
> MSFT: Microsoft Corp. (ÎßàÏù¥ÌÅ¨Î°úÏÜåÌîÑÌä∏) - Windows, Office, ÌÅ¥ÎùºÏö∞Îìú ÏÑúÎπÑÏä§(Azure)Î•º Ï†úÍ≥µÌïòÎäî ÏÜåÌîÑÌä∏Ïõ®Ïñ¥ Í∏∞ÏóÖ
>
> GOOGL: Alphabet Inc. (ÏïåÌååÎ≤≥) - GoogleÏùò Î™®ÌöåÏÇ¨Î°ú Í≤ÄÏÉâ ÏóîÏßÑ, YouTube, Android Îì±ÏùÑ Ïö¥ÏòÅÌïòÎäî Í∏∞ÏóÖ
>
> AMZN: Amazon.com Inc. (ÏïÑÎßàÏ°¥) - ÏÑ∏Í≥Ñ ÏµúÎåÄ Ï†ÑÏûêÏÉÅÍ±∞Îûò Í∏∞ÏóÖÏù¥Ïûê ÌÅ¥ÎùºÏö∞Îìú ÏÑúÎπÑÏä§(AWS) Ï†úÍ≥µ Í∏∞ÏóÖ
>
> META: Meta Platforms, Inc. (Î©îÌÉÄ) - Facebook, Instagram, WhatsApp Îì±ÏùÑ ÏÜåÏú†Ìïú ÏÜåÏÖú ÎØ∏ÎîîÏñ¥ Í∏∞ÏóÖ


#### **3Îã®Í≥Ñ: Ïû¨Î¨¥ Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞**
`yfinance`Ïùò `Ticker` Í∞ùÏ≤¥Î•º ÏÇ¨Ïö©ÌïòÏó¨ Ïû¨Î¨¥Ï†úÌëú Îç∞Ïù¥ÌÑ∞Î•º Î∂àÎü¨Ïò¨ Ïàò ÏûàÏäµÎãàÎã§. ÌäπÌûà Í∞êÏÇ¨Ïóê Ï§ëÏöîÌïú `ÎåÄÏ∞®ÎåÄÏ°∞Ìëú(Balance Sheet)`ÏôÄ `ÏÜêÏùµÍ≥ÑÏÇ∞ÏÑú(Income Statement)` Îç∞Ïù¥ÌÑ∞Î•º Í∞ÄÏ†∏Ïò§Îäî Í≤ÉÏù¥ ÌïµÏã¨ÏûÖÎãàÎã§.

````python
# Î∂àÎü¨Ïò¨ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Ï†ÄÏû•Ìï† ÎîïÏÖîÎÑàÎ¶¨
financial_data = {}

for ticker in tickers:
    try:
        # Í∞Å Í∏∞ÏóÖÏùò Ticker Í∞ùÏ≤¥ ÏÉùÏÑ±
        stock = yf.Ticker(ticker)

        # Ïû¨Î¨¥Ï†úÌëú Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
        # .income_stmt: ÏÜêÏùµÍ≥ÑÏÇ∞ÏÑú (Í≥ºÍ±∞ 5Í∞úÎÖÑ Îç∞Ïù¥ÌÑ∞)
        # .balance_sheet: ÎåÄÏ∞®ÎåÄÏ°∞Ìëú (Í≥ºÍ±∞ 5Í∞úÎÖÑ Îç∞Ïù¥ÌÑ∞)
        income_stmt = stock.income_stmt
        balance_sheet = stock.balance_sheet

        # ÌïÑÏöîÌïú Í≤ΩÏö∞, Îç∞Ïù¥ÌÑ∞Î•º Ï†ÑÏπò(transpose)ÌïòÏó¨ ÏãúÍ≥ÑÏó¥ ÌòïÌÉúÎ°ú ÎßåÎì¶
        # yfinanceÎäî ÏµúÏã† Ïû¨Î¨¥Ï†úÌëúÍ∞Ä Í∞ÄÏû• ÏôºÏ™ΩÏóê ÏúÑÏπò
        income_stmt = income_stmt.T
        balance_sheet = balance_sheet.T

        # Îç∞Ïù¥ÌÑ∞ Ï†ÄÏû•
        financial_data[ticker] = {
            'income_stmt': income_stmt,
            'balance_sheet': balance_sheet
        }

        print(f"{ticker}Ïùò Ïû¨Î¨¥ Îç∞Ïù¥ÌÑ∞Î•º ÏÑ±Í≥µÏ†ÅÏúºÎ°ú Î∂àÎü¨ÏôîÏäµÎãàÎã§.")

    except Exception as e:
        print(f"{ticker}Ïùò Îç∞Ïù¥ÌÑ∞Î•º Î∂àÎü¨Ïò§Îäî Ï§ë Ïò§Î•òÍ∞Ä Î∞úÏÉùÌñàÏäµÎãàÎã§: {e}")

# 'AAPL' Í∏∞ÏóÖÏùò ÏÜêÏùµÍ≥ÑÏÇ∞ÏÑú ÌôïÏù∏
financial_data['AAPL']['income_stmt'].head()
````

> ### üîç **Îç∞Ïù¥ÌÑ∞ ÏàòÏßë Í≥ºÏ†ïÏóê ÎåÄÌïú Ï∂îÍ∞Ä ÏÑ§Î™Ö**
>
> **1. `yfinance` Îç∞Ïù¥ÌÑ∞Îäî Î™á ÎÖÑÏπòÎ•º Î∂àÎü¨Ïò§ÎÇòÏöî?**
>
> `yfinance`Îäî Í∏∞Î≥∏Ï†ÅÏúºÎ°ú **Í∞ÄÏû• ÏµúÍ∑º 5Í∞úÎÖÑ**Ïùò Ïû¨Î¨¥ Îç∞Ïù¥ÌÑ∞Î•º Î∂àÎü¨ÏòµÎãàÎã§. ÌòÑÏû¨ ÎùºÏù¥Î∏åÎü¨Î¶¨Îäî Î∂àÎü¨Ïò¨ Í∏∞Í∞ÑÏùÑ ÏßÄÏ†ïÌïòÎäî Í∏∞Îä•ÏùÑ Í≥µÏãùÏ†ÅÏúºÎ°ú ÏßÄÏõêÌïòÏßÄ ÏïäÏïÑÏöî. ÎßåÏïΩ 5ÎÖÑ Ïù¥ÏÉÅÏùò Îç∞Ïù¥ÌÑ∞Î•º ÏõêÌïúÎã§Î©¥, ÏïÑÎûòÏôÄ Í∞ôÏùÄ Î∞©Î≤ïÏùÑ Í≥†Î†§Ìï† Ïàò ÏûàÏäµÎãàÎã§.
>
> * **Îç∞Ïù¥ÌÑ∞ ÏÜåÏä§ Î≥ÄÍ≤Ω**: `pandas_datareader` Í∞ôÏùÄ ÎùºÏù¥Î∏åÎü¨Î¶¨Îäî Îçî Í∏¥ Í∏∞Í∞ÑÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï†úÍ≥µÌï† Ïàò ÏûàÏäµÎãàÎã§.
> * **ÏàòÎèô Îç∞Ïù¥ÌÑ∞ Í≤∞Ìï©**: DART(Ï†ÑÏûêÍ≥µÏãúÏãúÏä§ÌÖú) Îì±ÏóêÏÑú ÏßÅÏ†ë Îç∞Ïù¥ÌÑ∞Î•º Îã§Ïö¥Î∞õÏïÑ `yfinance` Îç∞Ïù¥ÌÑ∞ÏôÄ Ìï©ÏπòÎäî Î∞©Î≤ïÏù¥ Í∞ÄÏû• Ï†ïÌôïÌï©ÎãàÎã§.
>
> **2. Ïôú Îç∞Ïù¥ÌÑ∞Î•º Ï†ÑÏπò(`Transpose`)Ìï¥Ïïº ÌïòÎÇòÏöî?**
>
> `yfinance`Í∞Ä Î∂àÎü¨Ïò® Îç∞Ïù¥ÌÑ∞Îäî Í∏∞Î≥∏Ï†ÅÏúºÎ°ú **ÏµúÏã† Ïó∞ÎèÑÍ∞Ä Í∞ÄÏû• ÏôºÏ™Ω**Ïóê ÏûàÍ≥†, **Ïó∞ÎèÑÍ∞Ä Ïó¥(column)**Ïóê, **Í∞Å ÏßÄÌëúÍ∞Ä Ìñâ(row)**Ïóê ÏúÑÏπòÌï©ÎãàÎã§. Ïù¥ ÌòïÌÉúÎäî Î∂ÑÏÑùÏóê Ï†ÅÌï©ÌïòÏßÄ ÏïäÏïÑÏöî.
>
> ```
> # Ï†ÑÏπò(Transpose) Ï†Ñ Îç∞Ïù¥ÌÑ∞ ÌòïÌÉú
>                     2024-12-31   2023-12-31  ...
> Operating Revenue        1000         950    ...
> Gross Profit              500         480    ...
> ```
>
> `df.T`Î•º ÏÇ¨Ïö©Ìï¥ **Îç∞Ïù¥ÌÑ∞Î•º Ï†ÑÏπò**ÌïòÎ©¥, **Í∞Å Ïó∞ÎèÑÍ∞Ä Ìñâ(row)**Ïù¥ ÎêòÍ≥†, **Í∞Å ÏßÄÌëúÍ∞Ä Ïó¥(column)**Ïù¥ Îê©ÎãàÎã§.
>
> ```
> # Ï†ÑÏπò(Transpose) ÌõÑ Îç∞Ïù¥ÌÑ∞ ÌòïÌÉú
>                 Operating Revenue  Gross Profit   Net Income ...
> 2024-12-31             1000          500          300
> 2023-12-31              950          480          280
> ```
>
> Ïù¥Î†áÍ≤å Î≥ÄÌôòÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÏùÄ Ïù¥Ï†êÏù¥ ÏûàÏäµÎãàÎã§.
>
> * **ÏãúÍ≥ÑÏó¥ Î∂ÑÏÑù Ïö©Ïù¥ÏÑ±**: Îç∞Ïù¥ÌÑ∞Ïùò ÏãúÍ∞Ñ ÌùêÎ¶ÑÏùÑ ÌïúÎààÏóê ÌååÏïÖÌï† Ïàò ÏûàÏñ¥ Ï∂îÏÑ∏ Î∂ÑÏÑùÏóê ÏµúÏ†ÅÌôîÎê©ÎãàÎã§.
> * **Ïû¨Î¨¥ ÎπÑÏú® Í≥ÑÏÇ∞ Ìé∏Î¶¨ÏÑ±**: Í∞Å Ïó∞ÎèÑÎ≥ÑÎ°ú ÏßÄÌëúÎ•º ÏâΩÍ≤å Í∞ÄÏ†∏ÏôÄÏÑú ÎπÑÏú®ÏùÑ Í≥ÑÏÇ∞Ìï† Ïàò ÏûàÏäµÎãàÎã§.
> * **Î®∏Ïã†Îü¨Îãù Î™®Îç∏Ïóê Ï†ÅÌï©**: ÎåÄÎ∂ÄÎ∂ÑÏùò Î™®Îç∏ÏùÄ Ìñâ(row)ÏùÑ ÌïòÎÇòÏùò Í¥ÄÏ∏°ÏπòÎ°ú Ïù∏ÏãùÌïòÍ∏∞ ÎïåÎ¨∏Ïóê, Ïù¥ ÌòïÌÉúÍ∞Ä Î™®Îç∏ ÌïôÏäµÏóê Îçî Ïú†Î¶¨Ìï©ÎãàÎã§.
>
> Í≤∞Î°†Ï†ÅÏúºÎ°ú, Îç∞Ïù¥ÌÑ∞Î•º Ï†ÑÏπòÌïòÎäî Í≤ÉÏùÄ Î∂ÑÏÑùÍ≥º Î™®Îç∏ÎßÅÏùò Î™©Ï†ÅÏóê ÎßûÍ≤å Îç∞Ïù¥ÌÑ∞ Íµ¨Ï°∞Î•º Ïû¨Íµ¨ÏÑ±ÌïòÎäî ÌïÑÏàòÏ†ÅÏù∏ Í≥ºÏ†ïÏûÖÎãàÎã§.

#### **4Îã®Í≥Ñ: Îç∞Ïù¥ÌÑ∞ ÌÜµÌï© Î∞è Ï†ÑÏ≤òÎ¶¨**
Î∂àÎü¨Ïò® Îç∞Ïù¥ÌÑ∞Î•º ÌïòÎÇòÎ°ú ÌÜµÌï©ÌïòÍ≥†, Í≤∞Ï∏°Í∞íÏù¥ÎÇò Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏùÑ Ï†ïÎ¶¨ÌïòÎäî Í≥ºÏ†ïÏûÖÎãàÎã§.

````python
# Î™®Îì† Í∏∞ÏóÖÏùò ÏÜêÏùµÍ≥ÑÏÇ∞ÏÑúÎ•º ÌÜµÌï© (Í∞ÑÎã®Ìïú ÏòàÏãú)
all_income_statements = pd.concat([data['income_stmt'] for data in financial_data.values()])

# Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨ (Í≤∞Ï∏°Í∞í Ï≤òÎ¶¨, Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ Î≥ÄÌôò Îì±)
# .fillna(0, inplace=True): Í≤∞Ï∏°Í∞íÏùÑ 0ÏúºÎ°ú Ï±ÑÏõÄ
# .astype(float): Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏùÑ Ïã§ÏàòÌòïÏúºÎ°ú Î≥ÄÌôò
all_income_statements.fillna(0, inplace=True)
all_income_statements = all_income_statements.astype(float)

print("Î™®Îì† Í∏∞ÏóÖÏùò ÏÜêÏùµÍ≥ÑÏÇ∞ÏÑú Îç∞Ïù¥ÌÑ∞Í∞Ä ÌÜµÌï© Î∞è Ï†ÑÏ≤òÎ¶¨ÎêòÏóàÏäµÎãàÎã§.")
````

> ####  üîç **Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨(Data Preprocessing)ÎûÄ?**
>
> Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨Îäî Î∂ÑÏÑùÏù¥ÎÇò Î™®Îç∏ÎßÅÏóê Ï†ÅÌï©ÌïòÎèÑÎ°ù Îç∞Ïù¥ÌÑ∞Î•º Í∞ÄÍ≥µÌïòÎäî Î™®Îì† Í≥ºÏ†ïÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§. Ïù¥Îäî ÌÜµÍ≥Ñ Î∂ÑÏÑùÏù¥ÎÇò AI Î™®Îç∏ Íµ¨Ï∂ïÏùò ÌïÑÏàò Îã®Í≥ÑÏòàÏöî.
>
> **Ïôú ÌïÑÏàòÏ†ÅÏù∏Í∞ÄÏöî?**
>
> * **Ï†ïÌôïÏÑ± ÌôïÎ≥¥**: ÎåÄÎ∂ÄÎ∂ÑÏùò ÏõêÏãú Îç∞Ïù¥ÌÑ∞ÏóêÎäî Ïò§Î•ò, ÎàÑÎùΩÎêú Í∞í(Í≤∞Ï∏°Í∞í), Ïù¥ÏÉÅÏπò Îì±Ïù¥ Ìè¨Ìï®ÎêòÏñ¥ ÏûàÏñ¥Ïöî. Ïù¥Îü∞ Îç∞Ïù¥ÌÑ∞Î•º Í∑∏ÎåÄÎ°ú ÏÇ¨Ïö©ÌïòÎ©¥ Î™®Îç∏Ïùò ÏòàÏ∏° Ï†ïÌôïÎèÑÍ∞Ä Îñ®Ïñ¥ÏßëÎãàÎã§. ÏòàÎ•º Îì§Ïñ¥, Ïû¨Î¨¥ Îç∞Ïù¥ÌÑ∞Ïóê ÎàÑÎùΩÎêú Í∞íÏù¥ ÎßéÏúºÎ©¥ Ï†ïÌôïÌïú Ïû¨Î¨¥ ÎπÑÏú®ÏùÑ Í≥ÑÏÇ∞Ìï† Ïàò ÏóÜÍ≤†Ï£†.
>
> * **Ìò∏ÌôòÏÑ± ÌôïÎ≥¥**: Î™®Îç∏Ïù¥ ÌïôÏäµÌï† Ïàò ÏûàÎäî ÌòïÌÉúÎ°ú Îç∞Ïù¥ÌÑ∞Î•º ÎßûÏ∂∞Ï§òÏïº Ìï¥Ïöî. ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î•º Ïà´ÏûêÎ°ú Î∞îÍæ∏Í±∞ÎÇò, Î≤îÏ£ºÌòï Îç∞Ïù¥ÌÑ∞Î•º Ïõê-Ìï´ Ïù∏ÏΩîÎî©ÌïòÎäî Îì±Ïùò ÏûëÏóÖÏù¥ Ïó¨Í∏∞Ïóê ÏÜçÌï©ÎãàÎã§.
>
> Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨ Í≥ºÏ†ïÏùÄ **"Garbage In, Garbage Out(Ïì∞Î†àÍ∏∞Í∞Ä Îì§Ïñ¥Í∞ÄÎ©¥ Ïì∞Î†àÍ∏∞Í∞Ä ÎÇòÏò®Îã§)"**Ïù¥ÎùºÎäî Îç∞Ïù¥ÌÑ∞ Í≥ºÌïôÏùò Í∏∞Î≥∏ ÏõêÏπôÏùÑ ÏßÄÌÇ§Îäî Í∞ÄÏû• Ï§ëÏöîÌïú Îã®Í≥ÑÏûÖÎãàÎã§.
