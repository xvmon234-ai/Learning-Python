```python
# 2.1 데이터 기본 정보 확인

# 2.1.1 데이터 불러오기
import pandas as pd

file_path = 'creditcard.csv'
df = pd.read_csv(file_path)

# 2.1.2 데이터 크기 및 구조 확인
print("----- df.shape -----")
print(df.shape)
print("\n ----- df.info() -----")
df.info()

print("\n ----- df.isnull().sum() -----")
null_counts = df.isnull().sum()
print(null_counts)

# 결측치 존재 여부를 확인하는 조건문
if null_counts.sum() == 0:
    print("\n[분석 내용]")
    print("모든 열에 결측치가 없음을 확인했습니다. 이는 추가적인 결측치 처리 과정 없이 바로 다음 단계로 넘어갈 수 있습니다.")
else:
    print("\n[분석 내용]")
    print("결측치가 존재하는 열이 있습니다. 모델 학습 전에 해당 결측치를 적절히 처리해야 합니다.")
    print("결측치 개수는 다음과 같습니다:")
    print(null_counts[null_counts > 0]) # 결측치가 있는 열만 출력

print("\n ----- df.describe() -----")
df.describe()
```

```python
# 2.2 EDA(탐색적 데이터 분석)

# 2.2.1 클래스 불균형 확인
import matplotlib.pyplot as plt
import seaborn as sns

print("\n ----- Class 불균형 확인 ----- ")
class_counts = df['Class'].value_counts()
print(class_counts)

print("\n ----- Class 불균형 Visualization ----- ")
sns.countplot(x='Class', data=df)

normal_count = class_counts[0]
fraud_count = class_counts[1]
total_count = len(df)

print(f"\n정상 거래(Class 0) 비율: {normal_count / total_count * 100:.4f}%")
print(f"이상 거래(Class 1) 비율: {fraud_count / total_count * 100:.4f}%")
print(f"총 데이터 개수: {total_count}")
```

```python
# 2.2.2 주요 피처 분포 시각화
normal = df[df['Class'] == 0]
fraud = df[df['Class'] == 1]

plt.figure(figsize=(10, 6))
sns.histplot(normal['Time'], bins=50, alpha=0.6, label='Normal')
sns.histplot(fraud['Time'], bins=50, color='red', alpha=0.6, label='Fraud')
plt.title('Time Distribution by Class', fontsize=15)
plt.xlabel('Time (in seconds)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend()
plt.show()
```

분석 결과:

이 그래프를 통해 정상 거래와 이상 거래의 Time 분포가 확연히 다르다는 패턴을 발견할 수 있습니다. 정상 거래는 하루 24시간 동안 고르게 분포하는 반면, 이상 거래는 특정 시간대(Time이 약 40,000초와 120,000초인 지점)에 집중적으로 발생했습니다. 이는 모델이 거래 발생 시간을 중요한 피처로 활용하여 이상 거래를 분류하는 데 핵심적인 단서가 될 것입니다.

```python
# 2.2.2 주요 피처 분포 시각화

plt.figure(figsize=(10, 6))
sns.histplot(normal['Amount'], bins=50, alpha=0.6, label='Normal')
sns.histplot(fraud['Amount'], bins=50, color='red', alpha=0.6, label='Fraud')
plt.title('Amount Distribution by Class', fontsize=15)
plt.xlabel('Amount', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend()
plt.show()
# Amount Feature에 대한 추가 분석이 필요하다고 판단돼 다음 셀에서 추가 분석을 수행
```

```python
# 2.2.2 주요 피처 분포 시각화
import numpy as np

# Amount 피처에 로그 변환 적용 (0을 방지하기 위해 1을 더함)
df['Log_Amount'] = np.log1p(df['Amount'])

# 로그 변환된 Amount 피처 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(df['Log_Amount'], bins=50)
plt.title('Log Transformed Amount Distribution', fontsize=15)
plt.xlabel('Log(Amount)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

분석 결과:

원래 Amount 피처는 대부분의 거래가 소액에 집중되어 있어 한쪽으로 치우친 분포를 보였습니다. 하지만 로그 변환을 적용한 결과, 데이터의 분포가 훨씬 안정적으로 변했습니다. 변환된 그래프를 통해 이상 거래는 정상 거래보다 더 다양한 금액대에서 발생하는 경향이 있음을 알 수 있습니다. 이는 단순히 금액의 크기만으로 이상 거래를 판단할 수 없으며, 복합적인 패턴을 학습해야 한다는 중요한 시사점을 제공합니다.

```python
# 2.3 데이터 전처리

# 2.3.1 Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['Scaled_Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df['Scaled_Time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))

# 원본 'Time'과 'Amount' 피처는 삭제합니다.
df = df.drop(['Time', 'Amount'], axis=1)
```

```python
# 2.3.2 Data Seperation
from sklearn.model_selection import train_test_split

X = df.drop('Class', axis=1) # 피처 데이터
y = df['Class'] # 타겟 데이터

# stratify 옵션을 사용하여 클래스 비율을 유지
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# 2.4 클래스 불균형 처리
# imbalanced-learn 라이브러리 설치
!pip install imbalanced-learn

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# 리샘플링 후의 데이터 분포 확인
print(f"Original shape: {y_train.shape}")
print(f"Resampled shape: {y_resampled.shape}")
print("Resampled class counts:")
print(y_resampled.value_counts())
```

```python
# 비교 분석을 위한 언더 샘플링 수행
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42)
X_rus, y_rus = rus.fit_resample(X_train, y_train)

print("RandomUnderSampler 적용 후 클래스 분포:\n", y_rus.value_counts())
```
