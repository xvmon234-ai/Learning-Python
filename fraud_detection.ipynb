```python
# 2.1 데이터 기본 정보 확인

# 2.1.1 데이터 불러오기
import pandas as pd

file_path = 'creditcard.csv'
df = pd.read_csv(file_path)

# 2.1.2 데이터 크기 및 구조 확인
print("----- df.shape -----")
print(df.shape)
print("\n ----- df.info() -----")
df.info()
print("\n ----- df.isnull().sum() -----")
df.isnull().sum()
print("\n ----- df.describe() -----")
df.describe()
```

```python
# 2.2 EDA(탐색적 데이터 분석)

# 2.2.1 클래스 불균형 확인
import matplotlib.pyplot as plt
import seaborn as sns

print("\n ----- Class 불균형 확인 ----- ")
df['Class'].value_counts()

print("\n ----- Class 불균형 Visualization ----- ")
sns.countplot(x='Class', data=df)
```

```python
# 2.2.2 주요 피처 분포 시각화
print("\n ----- Time 피처 분석 ----- ")
sns.histplot(df['Time'])
# issue: user warning 발생 > seaborn의 업데이트로 인해 distplot이 더 이상 사용 불가능할 예정이라는 의미의 warning, 이에 distplot을 histplot으로 변경


```

```python
# 2.2.2 주요 피처 분포 시각화
print("\n ----- Amount 피처 분석 ----- ")
sns.histplot(df['Amount'])
```

```python
# 2.3 데이터 전처리

# 2.3.1 Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['Scaled_Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
df['Scaled_Time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))

# 원본 'Time'과 'Amount' 피처는 삭제합니다.
df = df.drop(['Time', 'Amount'], axis=1)
```

```python
# 2.3.2 Data Seperation
from sklearn.model_selection import train_test_split

X = df.drop('Class', axis=1) # 피처 데이터
y = df['Class'] # 타겟 데이터

# stratify 옵션을 사용하여 클래스 비율을 유지
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# 2.4 클래스 불균형 처리
# imbalanced-learn 라이브러리 설치
!pip install imbalanced-learn

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# 리샘플링 후의 데이터 분포 확인
print(f"Original shape: {y_train.shape}")
print(f"Resampled shape: {y_resampled.shape}")
print("Resampled class counts:")
print(y_resampled.value_counts())
```
